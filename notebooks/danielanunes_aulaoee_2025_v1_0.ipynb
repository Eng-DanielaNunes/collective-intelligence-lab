```python
import plotly.graph_objects as go

# Dados fictícios para OEE
disponibilidade = 87.5
desempenho = 83.3
qualidade = 97.1

# Calcular OEE
OEE = (disponibilidade / 100) * (desempenho / 100) * (qualidade / 100) * 100

# Criando o gráfico de dashboard
fig = go.Figure()

# Adicionar barras para cada fator
fig.add_trace(go.Bar(
    x=["Disponibilidade", "Desempenho", "Qualidade"],
    y=[disponibilidade, desempenho, qualidade],
    name="Fatores OEE",
    marker=dict(color=['blue', 'orange', 'green']),
))

# Adicionar uma linha para o OEE total
fig.add_trace(go.Scatter(
    x=["OEE"],
    y=[OEE],
    mode='markers+text',
    name="OEE Total",
    marker=dict(size=20, color='red'),
    text=[f'{OEE:.2f}%'],
    textposition="top center",
))

# Layout do gráfico
fig.update_layout(
    title="Dashboard de OEE",
    xaxis_title="Fatores",
    yaxis_title="Porcentagem (%)",
    showlegend=False,
    template="plotly_dark",
    margin=dict(l=50, r=50, t=50, b=50),
    height=500
)

# Exibir o gráfico
fig.show()
```
