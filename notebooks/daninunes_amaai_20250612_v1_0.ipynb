```python
"""
# Projeto: AMA AI – Sistema de Apoio à Modelagem Automatizada
# Autora: Daniela Nunes dos Santos Ferreira
# Ano de Desenvolvimento: 2025
# Data: 12/06/2025
# Versão: 1.0
# Descrição: Gráficos de Gantt e métricas de gestão de projetos para o projeto AMA AI.

Descrição:
Este gráfico apresenta a linha do tempo das macroentregas do projeto AMA AI, que envolve a implementação de soluções de Inteligência Artificial para suporte à decisão em ambientes cívicos e industriais. O projeto inclui:
- Planejamento e levantamento de requisitos
- Mapeamento da arquitetura atual e avaliação de maturidade MCDM
- Desenvolvimento de roadmap de Arquitetura de Dados Moderna
- Implementação de Industrial Data Fabric
- Estudo e implantação de arquiteturas multiagente com LLMs
- Proof of Concept (PoC) e implantação de agentes de IA generativa
- Desenvolvimento de frontend para interação com usuários
- Avaliação da PoC e dos sistemas implementados
- Encerramento do projeto e entrega final


Métricas de Gestão de Projetos:
- Número de macroentregas: 9
- Duração total do projeto: 18 meses
- Atividades com sobreposição de execução: ME 4 & ME 5, ME 6 & ME 7, ME 8 & ME 6
- Macroentrega final destacada (ME 9) como marco de encerramento
- Duração média das macroentregas: {:.2f} meses
- Duração máxima da macroentrega: {} meses (ME {})
- Duração mínima da macroentrega: {} meses (ME {})

Este gráfico auxilia no acompanhamento do progresso do projeto, facilita a gestão do tempo, a identificação de atividades críticas, sobreposição de tarefas e comunicação com stakeholders.
"""

import matplotlib.pyplot as plt
import pandas as pd

# =======================
# Dados do Projeto AMA AI
# =======================
entregas = [
    ("ME 1 - Planejamento e Requisitos", 1, 2),
    ("ME 2 - Arquitetura Atual e Maturidade MCDM", 3, 6),
    ("ME 3 - Roadmap de Arquitetura de Dados", 6, 8),
    ("ME 4 - Implementação do Data Fabric", 9, 14),
    ("ME 5 - Estudo de Arquitetura Multiagente", 9, 11),
    ("ME 6 - PoC e Implantação do Multiagente IA", 12, 17),
    ("ME 7 - Desenvolvimento do Frontend", 13, 14),
    ("ME 8 - Avaliação da PoC e Data Fabric", 16, 17),
    ("ME 9 - Encerramento do Projeto", 18, 18),
]

# Criar DataFrame e calcular duração
df = pd.DataFrame(entregas, columns=["Atividade", "Início", "Fim"])
df["Duração"] = df["Fim"] - df["Início"] + 1

# =======================
# Função para plotar Gantt
# =======================
def plot_gantt(df):
    fig, ax = plt.subplots(figsize=(12, 6))
    colors = ["#0F4C75", "#3282B8"]

    for i, row in df.iterrows():
        cor = "#F9A825" if row["Atividade"].startswith("ME 9") else colors[i % 2]
        ax.barh(
            y=row["Atividade"],
            width=row["Duração"],
            left=row["Início"],
            color=cor,
            edgecolor='gray'
        )

    ax.set_xlabel("Mês de Execução")
    ax.set_ylabel("Macroentregas")
    ax.set_title("Gráfico de Gantt do Projeto AMA AI")
    ax.set_xlim(1, 19)
    ax.set_xticks(range(1, 19))
    plt.gca().invert_yaxis()
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# =======================
# Função para plotar métricas
# =======================
def plot_metrics(df):
    duracao_media = df["Duração"].mean()
    duracao_max = df["Duração"].max()
    atividade_max = df.loc[df["Duração"].idxmax(), "Atividade"]
    duracao_min = df["Duração"].min()
    atividade_min = df.loc[df["Duração"].idxmin(), "Atividade"]

    # Dados para gráfico
    metricas = pd.DataFrame({
        "Métrica": ["Duração Média", "Duração Máx.", "Duração Mín."],
        "Meses": [duracao_media, duracao_max, duracao_min],
        "Atividade": ["Todas", atividade_max, atividade_min]
    })

    fig, ax = plt.subplots(figsize=(8, 4))
    bars = ax.bar(metricas["Métrica"], metricas["Meses"], color=["#3282B8", "#0F4C75", "#F9A825"])

    # Anotação das atividades
    for bar, atividade in zip(bars, metricas["Atividade"]):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, atividade,
                ha='center', va='bottom', fontsize=10)

    ax.set_ylabel("Meses")
    ax.set_title("Métricas de Duração das Macroentregas")
    plt.tight_layout()
    plt.show()

# =======================
# Executar plots
# =======================
plot_gantt(df)
plot_metrics(df)


```
